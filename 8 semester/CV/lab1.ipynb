```
from PIL import Image
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
```

```
SHIFTING = 15
PATH = 'image.jpg'
```

```
def show_hist(image: Image):
    img.show()
    display(img)

    image = image.convert('L')
    plt.plot(list(range(256)), image.histogram())
    plt.show()
```

```
img = Image.open(PATH)
img.load()
show_hist(img)
```

```
data = np.array(img)

for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        for k in range(3):
            pixel = data[i][j][k] - SHIFTING
            if pixel < 0:
                data[i][j][k] = 0
            else:
                data[i][j][k] = pixel
```

```
img = Image.fromarray(data, 'RGB')
show_hist(img)
```
